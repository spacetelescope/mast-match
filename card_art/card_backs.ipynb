```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import numpy as np
from pandas import read_csv
import matplotlib.patheffects as pe
from matplotlib.patches import FancyBboxPatch, BoxStyle
from matplotlib.colors import LinearSegmentedColormap
```

# Observation Card Backs

```python
mast_logo = Image.open(f'icons/mast_logo.jpg').resize((1400,1400))

# Change MAST logo to transparent - only get black and white pixels
mast_logo_array = np.copy(np.asarray(mast_logo.getdata()).reshape(mast_logo.size[0], mast_logo.size[1], 3))
rgba_array = np.asarray(np.copy(mast_logo.convert('RGBA')))

for i in range(mast_logo_array.shape[0]):
    for j in range(mast_logo_array.shape[1]):
        r,g,b,a = rgba_array[i,j]
        if (b) <= 50: # if black = opaque
            a = 255
        else:
            a = 0
            
        if r >= 100: # if white = opaque
            a = 255
            rgba_array[i,j] = [255,255,255,a]
        rgba_array[i,j][3] = a
        
```

```python
fig, ax = plt.subplots()
fig.set_size_inches((2.5,3.5))

#img = Image.open(f'obs_card_imgs/carina.png')
#img = img.resize((1000,1400)) #card size is 2.5 x 3.5 inches
#ax.imshow(img,zorder=0, extent=(0,1000,0,1400))

#cmap = LinearSegmentedColormap.from_list("", ['royalblue','deepskyblue',])
cmap = LinearSegmentedColormap.from_list("", 
               ["black","midnightblue", "#1a619f",
                "deepskyblue","white"])
#cmap = LinearSegmentedColormap.from_list("", ['indigo', '#1a619f','deepskyblue',])
plt.imshow([[0,0.1], [0.5,1.0]], 
            cmap = cmap, interpolation = 'bicubic',extent=(0,1000,0,1400))

    
# Game title
#ax.text(75, 1300, 'MAST\nMatch', ha='left', va='top', color='k',
#             fontsize=16, font='silom', fontweight='bold',
#        path_effects=[pe.withStroke(linewidth=2, foreground="w")])


# Transparent MAST Logo
x,y = 500,700 # center of image
size=900
ax.imshow(rgba_array,zorder=2,extent=(x-size,x+size,y-size,y+size))


# Type of Card
stripe = plt.Rectangle((0,100), 1200, 150,color='k')
ax.add_patch(stripe)
ax.text(500, 175, 'Observation Card', ha='center', va='center', color='white',
             fontsize=12, font='silom', fontweight='bold')




# Border Around Edge
r = plt.Rectangle((0, 0), 1000, 1400, zorder=10, facecolor='none', edgecolor='k', lw=10)
ax.add_patch(r)
roundbox = FancyBboxPatch((0, 0), 1000, 1400,
                          boxstyle=BoxStyle("Round",rounding_size=200),
                          facecolor='none', edgecolor='k', lw=10,zorder=10)
ax.add_patch(roundbox)

# Additional Formatting
ax.set_xlim(0,1000)
ax.set_ylim(0,1400)
ax.set_axis_off()
plt.axis('off')

plt.savefig(f'final_cards/card_backs/BACK_observation.png', bbox_inches='tight', pad_inches=0.01, dpi=300)
plt.show()
    
```
